```python
from tensorflow import keras
import numpy as np

# Load the Keras model
model = keras.models.load_model('sentiment_model.keras')
print('âœ“ BiLSTM Model loaded successfully!')
print('\nModel Summary:')
model.summary()
```

```python
# Test with your review - change the text and run
import tensorflow as tf

review = "bonne annÃ©e ana ou  houwa"

# Convert to TensorFlow string tensor
text_input = tf.constant([review], dtype=tf.string)

# Make prediction
result = model.predict(text_input, verbose=0)[0]

# Map predictions to sentiment labels
sentiments = {0: "Negative", 1: "Neutral", 2: "Positive"}
predicted_class = np.argmax(result)
predicted_sentiment = sentiments[predicted_class]
confidence = result[predicted_class]

print("=" * 70)
print(f"ğŸ“ Review: {review}")
print("=" * 70)
print(f"\nğŸ¯ Predicted Sentiment: {predicted_sentiment}")
print(f"ğŸ“Š Confidence: {confidence:.2%}")
print(f"\nğŸ“ˆ Probability Distribution:")
print(f"   Negative: {result[0]:.4f} ({result[0]*100:.2f}%)")
print(f"   Neutral:  {result[1]:.4f} ({result[1]*100:.2f}%)")
print(f"   Positive: {result[2]:.4f} ({result[2]*100:.2f}%)")
print("=" * 70)
```
